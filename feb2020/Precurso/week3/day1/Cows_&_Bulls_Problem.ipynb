Create a program that will play the “cows and bulls” game with the user. The game works like this:

Randomly generate a 4-digit number. Ask the user to guess a 4-digit number. For every digit that the user guessed correctly in the correct place, they have a “cow”. For every digit the user guessed correctly in the wrong place is a “bull.” Every time the user makes a guess, tell them how many “cows” and “bulls” they have. Once the user guesses the correct number, the game is over. Keep track of the number of guesses the user makes throughout teh game and tell the user at the end.

Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  Enter a number: 
  >>> 1234

  2 cows, 0 bulls
  
  >>> 1256

  1 cow, 1 bull
  
  ...

5921

    -1295 --> 4 BULLS y 0 cows

    -3333 --> 0 cows y 0 bulls

    -1221 --> 2 bulls y 2 cows

```python
import random

# Generar número random
def random_generate(digits):
    """ Generate a random with 'digits' digits"""
    return random.randint(0, digits)

x = "%4d" % random_generate(digits=9999)
print(x)
print(type(x))
```

```python
#user_number = input("Insert your number:")

# Bulls: 1
# Cows: 2 
to_check = "6822"
user_number = "6772"
```

```python
def check_chars(char1, char2):
    if char1 == char2:  # 'char1' and 'char2' are equals.
        return True
    else:
        return False
```

```python
def check_numbers_and_return_cows_bulls_numbers(user, check_machine):
    """ This function checks the conditions ... """
    cows_number = 0
    bulls_number = 0
    for i, char in enumerate(check_machine):
        if char in user:  # if True, then one bull or cow
            if check_chars(char, user[i]):
                cows_number += 1
            else:
                bulls_number += 1
    return cows_number, bulls_number
```

```python
cows, bulls = check_numbers_and_return_cows_bulls_numbers(user=user_number, check_machine=to_check)

print("Cows: ", cows)
print("Bulls: ", bulls)
```
